9. Escribe un programa que, usando el módulo random y el módulo statistics, establezca la semilla pseudoaleatoria y posteriormente genere 100 muestras de la distribución normal. A continuación, debe imprimir por pantalla la media y la desviación tipica de esas muestras.

```python
import statistics
```

```python
random.seed(43)
muestras = [random.gauss() for _ in range(100)]

media_obtenida = statistics.mean(muestras)
desv_tipica_obtenida = statistics.stdev(muestras)

print(f"La media de las muestras es {media_obtenida}")
print(f"La desviación típica de las muestras es {desv_tipica_obtenida}")
```

10. Escribe un programa que, usando el módulo random y el módulo statistics, establezca la semilla pseudoaleatoria y posteriormente genere 100 muestras de la distribución uniforme entre 3 y 7. A continuación, debe imprimir por pantalla la mediana de esas muestras, así como los deciles de dichas muestras.

```python
random.seed(553)
muestras = [random.uniform(3, 7) for _ in range(100)]

mediana_obtenida = statistics.median(muestras)

print(f"La mediana de las muestras es {media_obtenida}")

deciles = statistics.quantiles(muestras, n = 10)

print(f"Los deciles de la muestra son: {deciles}")
```
