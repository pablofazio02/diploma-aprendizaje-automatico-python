1. Escribe una función que calcule el factorial de un número entero dado.

```python
def factorial (num):
  if num == 0:
    return 1
  else:
    return num * factorial(num - 1)
```

```python
factorial(10)
```

2. Escribe una función para calcular la suma de todos los elementos de una lista.

```python
def suma_lista(lista_ejemplo):
  suma = 0

  for elem in lista_ejemplo:
    suma += elem

  return suma
```

```python
suma_lista([7,4,8,3,5,76,3,46,38])
```

3. Escribe una función para contar el número de vocales de una cadena.

```python
def contar_vocales(cadena_ejemplo):
  cont = 0

  for caracter in cadena_ejemplo:
    if caracter in "AEIOUaeiouáéíóúÁÉÍÓÚ":
      cont += 1

  return cont
```

```python
contar_vocales("hbchyfyeunsjwuianijoxn")
```

4. Escribe una función para comprobar si una cadena es un palíndromo.

```python
def palindromo (cadena_param):
  return cadena_param == cadena_param[::-1]
```

```python
palindromo("reconocer")
```

5. Escribe una función que elimine los duplicados de una lista.

```python
def elim_duplicado (lista_param):
  conjunto = set(lista_param)
  return list(conjunto)
```

```python
elim_duplicado([4,7,3,8,5,9,4,3,6,7,8])
```

6. Escribe una función que calcula el mínimo y el máximo elemento de una lista, y los devuelve en una tupla.

```python
def calc_min_max (lista_ej):
  min = lista_ej[0]
  max = lista_ej[0]

  for i in range(1, len(lista_ej)):
    if min > lista_ej[i]:
      min = lista_ej[i]
    if max < lista_ej[i]:
      max = lista_ej[i]

  return min, max
```

```python
calc_min_max([4,6,8,3,-1,8,9,45,3])
```

7. Escribe una función para comprobar si un número es primo.

```python
def esPrimo (n):
  cont = 0
  for i in range(2, n):
    if n % i == 0:
      return False

  return True
```

```python
esPrimo(101)
```

9. Escribe una función para hallar el segundo menor elemento de una lista de números.

```python
def segundo_menor (lista):
  lista_aux = sorted(lista)
  return lista_aux[1]
```

```python
segundo_menor([3,5,3,2,9,6,5,-1,0,4,-78])
```

10. Escribe una función para comprobar si dos cadenas son anagramas.

```python
def son_anagramas (cadena1, cadena2):
  cadena_aux1 = sorted(cadena1)
  cadena_aux2 = sorted(cadena2)
  return cadena_aux1 == cadena_aux2
```

```python
son_anagramas("holah", "aloha")
```

11. Escribe una función para calcular la suma de todos los números pares que haya en una lista.

```python
def suma_pares(listaparam):
  suma = 0
  for elem in listaparam:
    if elem % 2 == 0:
      suma += elem
  return suma
```

```python
suma_pares([3,6,7,9,4,12,987,45,90])
```

12. Escribe una función para convertir una lista de cadenas a mayúsculas.

```python
def mayusculas (lista_cadenas):
  lista_res = []
  for cadena in lista_cadenas:
    lista_res.append(cadena.upper())
  return lista_res
```

```python
mayusculas(["hola", "pegAA", "como", "rasO"])
```
