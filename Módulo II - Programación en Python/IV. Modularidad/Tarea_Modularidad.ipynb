1. Escribe una función denominada *“calcular_estadisticas”* para calcular estadísticas básicas de una lista de números (mínimo, máximo, promedio). Usa argumentos posicionales, el primer argumento será la lista de números, el segundo indicará si incluir números negativos en el cálculo (por defecto sí) y el tercer argumento indicará el tipo de operación estadística (por defecto, calcular el
“mínimo”). Considera la esta lista de números `[5, -3, 8, 15, -2, 9]` y haz lo siguiente:

    **a)** Calcula la salida para la lista de números utilizando los argumentos por defecto.

    **b)** Calcula el mínimo de la lista de números sin incluir negativos, pero invoca a la función con los argumentos en un orden diferente al que has usado al declarar la función.



```python
# Importamos la función mean() del módulo 'statistics' para usarla en la función
from statistics import mean
```

```python
def calcular_estadisticas(lista, incluir_negativos=True, tipo_operacion="mínimo"):

  """
    Calcula estadísticas básicas de una lista de números.

    Args:
        lista (list): Lista de números (int o float).
        incluir_negativos (bool): Si es False, excluye los números negativos del cálculo; Valor por defecto = True
        tipo_operacion (str): Operación a realizar ("mínimo", "promedio", "máximo"); Valor por defecto = "mínimo"

    Returns:
        float/int: El resultado de la operación.
        value Error: Si la operación no es válida o la lista esta vacía.
    """

  if not incluir_negativos:
  # Filtramos los negativos si el parámetro está en False
      lista = [x for x in lista if x >= 0]

  if not lista:
    raise ValueError("Error: La lista final está vacía")

  if tipo_operacion == "mínimo":
      return min(lista)
  elif tipo_operacion == "promedio":
      return mean(lista)
  elif tipo_operacion == "máximo":
      return max(lista)
  else:
      raise ValueError("Error: Tipo de operación no válida")
```

```python
# APARTADO A)

lista_entrada = [5, -3, 8, 15, -2, 9]
calcular_estadisticas(lista_entrada) # Se usan los argumentos por defecto
```

```python
# APARTADO B)

# Observemos que hemos modificado el orden de paso de parámetros pero por ello
# necesitamos pasarles su nombre de argumento

calcular_estadisticas(incluir_negativos=False, tipo_operacion="mínimo", lista = lista_entrada)
```

3. Define una función generadora denominada *genera_cuadrados*, haciendo uso de la sentencia *yield*, que genere una secuencia infinita de números cuadrados partiendo desde el 0. Después usa esa función en un bucle que vaya imprimiendo por pantalla cuadrados hasta que llegue a un cuadrado igual o mayor que 50, en ese momento se detendrá el bucle.


```python
def genera_cuadrados():
  """
  Generador infinito que devuelve el cuadrado de números consecutivos
  empezando desde 0.
  """
  i = 0
  while True: # -> Bucle infinito
    yield i**2
    i += 1
```

```python
cuadrado = genera_cuadrados()
cuadrado_siguiente = 0

while cuadrado_siguiente < 50:
  # Como hemos usado yield, next() retorna el siguiente valor del generador
  cuadrado_siguiente = next(cuadrado)
  print(cuadrado_siguiente)
```

```python
from statistics import mean

import math
```

```python
class AnalisisDatos():

  """
    Clase base para realizar análisis estadísticos básicos sobre una lista de datos.
  """

  def __init__ (self, lista_inicial):
    self.lista = lista_inicial

  def calcular_desviacion_estandar(self):
    """
      Calcula la desviación típica de los datos.
      Fórmula: sqrt(sum((x - media)^2)/N)
    """
    media_aritmetica = mean(self.lista)
    num_elems = len(self.lista)
    suma = 0
    for elem in self.lista:
      suma += (elem - media_aritmetica)**2

    return math.sqrt(suma/num_elems)
```

```python
class AnalisisExtendido(AnalisisDatos):

  """
    Clase hija que hereda de AnalisisDatos e incluye
    el cálculo de la varianza.
  """

  # No es necesario definir __init__ de nuevo, se usará el de la clase padre

  def calcular_varianza(self):
    """
      Calcula la varianza de los datos.
      Fórmula: sum((x - media)^2)/N
    """
    media_aritmetica = mean(self.lista)
    num_elems = len(self.lista)
    suma = 0
    for elem in self.lista:
      suma += (elem - media_aritmetica)**2

    return (suma/num_elems)
```

```python
# APARTADO A)

lista_ejemplo = [10, 12, 23, 23, 16, 23, 21, 16]

analisis = AnalisisExtendido(lista_ejemplo)

print(f"La desviación típica de los datos de la lista es {analisis.calcular_desviacion_estandar()}")

print(f"La varianza de los datos de la lista es {analisis.calcular_varianza()}")
```

5.  Crea una clase llamada *Complejo* que implemente los números complejos en representación binomial, $z = a + bi$, donde “$a$” es la parte real y “$b$” es la parte imaginaria del número complejo $z$. Implementa la suma, resta, producto, cociente de números complejos y también el método mágico que imprime como cadena, todo ello sobrecargando los operadores correspondientes.


$$
(a + bi) + (c + di) = (a + c) + (b + d)i \\
(a + bi) - (c + di) = (a - c) + (b - d)i \\
(a + bi) \cdot (c + di) = (ac - bd) + (ad + bc)i \\
\frac{(a+ bi)}{(c + di)} = \frac{(ac + bd)}{(c^2 + d^2)} + \frac{(bc - ad)}{(c^2 + d^2)}i
$$

Prueba el correcto funcionamiento de la clase y todas las operaciones implementadas, imprimiendo por consola los resultados obtenidos. Compare dichos resultados con los obtenidos con la clase predefinida *cmath.complex*

```python
class Complejo():
  """
  Implementa números complejos en representación binomial (z = a + bi).

  Esta clase permite realizar operaciones aritméticas básicas (suma, resta,
  multiplicación y división) sobrecargando los operadores nativos de Python.
  """

  def __init__ (self, a=0, b=0):
    """
    Constructor de la clase Complejo.

    Args:
        a (float/int): Parte real del número complejo. Por defecto es 0.
        b (float/int): Parte imaginaria del número complejo. Por defecto es 0.
    """
    self.a = a
    self.b = b

  def __str__(self):
    """
    Devuelve la representación en cadena del número complejo.
    Formato: (a+bj), (a-bj) ó bj.
    """
    if self.a == 0:
      return f"{self.b}j"
    else:
      if self.b >= 0:
        return f"({self.a}+{self.b}j)"
      else:
        return f"({self.a}{self.b}j)"

  def __add__(self, other):
    """
    Sobrecarga del operador suma (+).
    Fórmula: (a + bi) + (c + di) = (a + c) + (b + d)i
    """
    new_a = self.a + other.a
    new_b = self.b + other.b
    return Complejo(new_a, new_b)

  def __sub__(self, other):
    """
    Sobrecarga del operador resta (-).
    Fórmula: (a + bi) - (c + di) = (a - c) + (b - d)i
    """
    new_a = self.a - other.a
    new_b = self.b - other.b
    return Complejo(new_a, new_b)

  def __mul__(self, other):
    """
    Sobrecarga del operador multiplicación (*).
    Fórmula: (ac - bd) + (ad + bc)i
    """
    new_a = self.a*other.a - self.b*other.b
    new_b = self.a*other.b + self.b*other.a
    return Complejo(new_a, new_b)

  def __truediv__(self, other):
    """
    Sobrecarga del operador división (/).
    Se realiza multiplicando numerador y denominador por el conjugado del denominador.

    Returns:
        Complejo: Resultado de la división.
        ZeroDivisionError: Si el denominador es 0.
    """
    denominador = other.a**2 + other.b**2
    if denominador == 0:
        raise ZeroDivisionError("Error: División por cero")
    new_a = (self.a*other.a + self.b*other.b) / denominador
    new_b = (self.b*other.a - self.a*other.b) / denominador
    return Complejo(new_a, new_b)
```

```python
complejo_prueba = Complejo(7.25, -2)
complejo_prueba2 = Complejo(8)
complejo_prueba3 = Complejo(0, 4.2)

print(f"La suma de {complejo_prueba} y {complejo_prueba3} es: {complejo_prueba + complejo_prueba3}")

print(f"La resta de {complejo_prueba2} y {complejo_prueba3} es: {complejo_prueba2 - complejo_prueba3}")

print(f"La multiplicación de {complejo_prueba2} y {complejo_prueba} es: {complejo_prueba * complejo_prueba2}")

print(f"La división de {complejo_prueba} y {complejo_prueba3} es: {complejo_prueba / complejo_prueba3}")
```

6. Crea una clase *ComplejoExtendido* que herede de la clase *Complejo* desarrollada en el ejercicio anterior. La clase *ComplejoExtendido* debe tener un método de clase que permita crear una instancia de la clase pasándole como argumentos el módulo y la fase del número complejo en forma polar. Dicho constructor debe mostrar un error si los argumentos que se le pasan son incorrectos, es decir, si el módulo es negativo. La clase debe implementar un método para imprimir como cadena de texto el número complejo en forma polar. Por último, la clase debe implementar métodos para hallar el módulo $|z|$ y la $fase(z)$:

$$
|z| = \sqrt{a^2+b^2}\\
fase(z) = arctan(b/a)
$$

**Pista:** debes redefinir los métodos para los operadores aritméticos. Prueba el correcto funcionamiento de la clase y todas las operaciones implementadas, imprimiendo por consola los resultados obtenidos. Compare dichos resultados con los obtenidos con la clase predefinida *cmath.complex*

```python
class ComplejoExtendido(Complejo):
  """
  Extensión de la clase Complejo para soportar operaciones y representaciones
  en formato polar.
  """

  @classmethod
  def desde_polar(cls, modulo, fase):
      """
      Constructor alternativo para crear una instancia
      directamente desde coordenadas polares.

      Args:
          modulo (float): La longitud del vector (r).
          fase (float): El ángulo en radianes (theta).

      Returns:
          ComplejoExtendido: Una nueva instancia inicializada.

      Raises:
          ValueError: Si el módulo es negativo.
      """
      if modulo < 0:
          raise ValueError("Error: El módulo no puede ser negativo")

      # Convertimos polar a binomial (a + bi) para guardarlo en la clase padre
      a = modulo * math.cos(fase)
      b = modulo * math.sin(fase)

      # Devolvemos una instancia de la clase llamando al constructor 'cls'
      return cls(a, b)

  def obtener_modulo(self):
    """
    Calcula el módulo del número complejo.
    Fórmula: |z| = sqrt(a^2 + b^2)
    """
    return math.sqrt(self.a**2 + self.b**2)

  def obtener_fase(self):
    """
    Calcula la fase/argumento del número complejo en radianes.
    Fórmula: fase(z) = arctan(b/a)
    """

    # Usamos math.atan2(y, x) en lugar de atan(y/x) para manejar el signo de los cuadrantes
    return math.atan2(self.b, self.a)

  def mostrar_polar(self):
    """
    Devuelve una cadena representando el número en forma polar:
    'r * (cos(theta) + i*sin(theta))'
    """
    r = self.obtener_modulo()
    theta = self.obtener_fase()
    return f"{r} * (cos({theta}) + i*sin({theta}))"

  # Redefinición de los operadores aritméticos
  def __add__(self, other):
    res = super().__add__(other) # Usamos la lógica del padre
    return ComplejoExtendido(res.a, res.b) # Convertimos el resultado al hijo

  def __sub__(self, other):
    res = super().__sub__(other)
    return ComplejoExtendido(res.a, res.b)

  def __mul__(self, other):
    res = super().__mul__(other)
    return ComplejoExtendido(res.a, res.b)

  def __truediv__(self, other):
    res = super().__truediv__(other)
    return ComplejoExtendido(res.a, res.b)
```
