Considera la siguiente definición de clase:


```python
class Empleado:
  def __init__(self, nombre, apellido1, apellido2, dni, cargo, departamento, sueldo):
    self.nombre = nombre
    self.apellido1 = apellido1
    self.apellido2 = apellido2
    self.dni = dni
    self.cargo = cargo
    self.departamento = departamento
    self.sueldo = sueldo
  def __str__(self):
    return f' Empleado {self.nombre} {self.apellido1} {self.apellido2} con DNI {self.dni}'

  def nombre_completo(self):
    return f'{self.apellido1} {self.apellido2}, {self.nombre}'

  def trasladar(self, nuevo_dept):
    self.departamento = nuevo_dept
```

4. Dada una lista de empleados, ordénala según sus nombres completos e imprímela por pantalla.

```python
empleado1 = Empleado(
    "Laura",
    "García",
    "Fernández",
    "12345678A",
    "Desarrolladora Backend",
    "Tecnología",
    45000
)

empleado2 = Empleado(
    "Carlos",
    "Rodríguez",
    "Ruiz",
    "87654321B",
    "Director Comercial",
    "Ventas",
    55000
)

empleado3 = Empleado(
    "Ana",
    "Martínez",
    "López",
    "11223344C",
    "Técnica de Selección",
    "Recursos Humanos",
    28000
)

empleado4 = Empleado(
    "David",
    "Sánchez",
    "Gómez",
    "55667788D",
    "Jefe de Almacén",
    "Logística",
    24000
)

empleado5 = Empleado(
    "Elena",
    "Pérez",
    "Jiménez",
    "99887766E",
    "Diseñadora Gráfica",
    "Marketing",
    30000
)

lista_empleados = [empleado1, empleado2, empleado3, empleado4, empleado5]
```

```python
lista_nombres_empleados = []

for elem in lista_empleados:
  lista_nombres_empleados.append(elem.nombre_completo())

lista_nombres_empleados.sort()

print(lista_nombres_empleados)
```

5. Dada una lista de empleados, calcula la lista de los que cobran más de 40000 e imprímela por pantalla.

```python
lista_sueldo = []

for elem in lista_empleados:
  if elem.sueldo > 40000:
    lista_sueldo.append(elem)

for i in lista_sueldo:
  print(i)
```

6. Dada una lista de empleados, aplícales una subida de sueldo del 20% y calcula la suma de todos los sueldos resultantes.

```python
suma = 0

for elem in lista_empleados:
  sueldo_anterior = elem.sueldo
  elem.sueldo = sueldo_anterior + 0.2*sueldo_anterior
  suma += elem.sueldo

print(f"Suma de todos los sueldos resultantes es: {suma} euros.")
```
