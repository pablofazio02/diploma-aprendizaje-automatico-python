1.  Vamos a calcular la hipotenusa de un triángulo rectángulo, para ello escribe un código que realice lo siguiente:

    a) Leer la magnitud del cateto mayor y del cateto menor. Asegúrese de que los valores introducidos sean flotantes.
    
    b) Muestre por pantalla una o varias cadenas de texto como se muestra a continuación:
    -  Comprobando que el cateto mayor sea mayor que cero: ***comprobacion1***
    - Comprobando que el cateto menor sea mayor que cero: ***comprobacion2***
    - Comprobando que el cateto mayor sea de mayor magnitud que el cateto menor: ***comprobacion3***
  
    donde ***comprobacion1*** es una variable que será **True** si el cateto mayor es mayor que cero, **False** en caso contrario, ***comprobacion2*** será **True** si el cateto menor es mayor que cero y ***comprobacion3*** será **True** si el cateto mayor es mayor que el cateto menor.
 - Calcule la magnitud de la hipotenusa como: *h* = $\sqrt{C^2 + c^2}$ donde *h* es la hipotenusa, *C* el cateto mayor y *c* el cateto menor.

````python
# APARTADO A)

cateto_mayor = float(input("Introduzca la magnitud del cateto mayor: "))
cateto_menor = float(input("Introduzca la magnitud del cateto menor: "))

# Basta usar la instrucción input teniéndo cuidado de realizar un 'casting' con float ya que nos devuelve valores 'str' y nosotros queremos hacer cálculos con estos valores.
````

````python
# APARTADO B)

# Redactamos las comprobaciones requeridas usando el símbolo '>' (mayor estricto)
comprobacion1 = cateto_mayor > 0
comprobacion2 = cateto_menor > 0
comprobacion3 = cateto_mayor > cateto_menor

# Imprimimos por pantalla las comprobaciones como en el ejemplo de ejecución
print("- Comprobando que el cateto mayor sea mayor que cero:", comprobacion1)
print("- Comprobando que el cateto menor sea mayor que cero:", comprobacion2)
print("- Comprobando que el cateto mayor sea mayor que el cateto menor:", comprobacion3)
````

````python
# Para el cálculo de la magnitud de la hipotenusa necesitamos usar el exponente '**2' y la función sqrt (de la librería math),
# luego, en primer lugar, importemos dicha función

from math import sqrt

h = sqrt(cateto_mayor**2 + cateto_menor**2)

print("La  hipotenusa es:", h)
````

4. Vamos a realizar algunos cálculos numéricos con variables de tipo entero y flotante:

    **a)** Aproxima el valor del número $e$, es decir, la base de los logaritmos naturales. Para ello, utiliza la siguiente aproximación:

    $$e \sim \left(1+\frac{1}{n}\right)^{n}$$

    siendo $n$ un número entero positivo grande. Debes usar un valor de $n$ lo suficientemente grande como para que el valor absoluto del error de la aproximación de $e$ que consigas sea inferior a $10^{-4}$. Puedes tomar como valor de referencia $e = 2.7182818284590452353602874713527$

   > **Nota:** El valor absoluto del error de la aproximación se calcula como $|e-e_{aprox}|$ donde $| \cdot |$ es el valor absoluto cuya definición es:
   >
   > $$|a| = \left\{ \begin{array}{ll} a & \text{si } a \ge 0 \\ -a & \text{en otro caso} \end{array} \right.$$

    **b)** Aproxima el valor del número $\pi$. Para ello, utiliza la aproximación de Madhava-Leibniz:

    $$\pi \sim 4\left(1-\frac{1}{3}+\frac{1}{5}-\frac{1}{7}+\dots\right)$$

    Debes usar una cantidad de sumandos lo suficientemente grande como para que el valor absoluto del error de la aproximación de que consigas sea inferior a $0.1$. Puedes tomar como valor de referencia $\pi = 3.14159265359$

    **c)** Lee un número en coma flotante $x$ por teclado, y calcula su primer dígito significativo $m$. Para ello, debes usar la siguiente fórmula:

    $$m = \lfloor \frac{|x|}{10^{\lfloor \log_{10}(|x|) \rfloor}} \rfloor$$

    siendo $| \cdot |$ el valor absoluto, $\log_{10}$ el logaritmo en base 10, y $\lfloor \cdot \rfloor$ el redondeo hacia abajo. Antes de hacer el cálculo debes insertar la siguiente línea en tu programa Python:

    `import math`

    De esta forma, podrás usar las siguientes funciones matemáticas:
     * `math.log10` para calcular el logaritmo en base 10.
     * `math.floor` para calcular el redondeo hacia abajo.

    **d)** Dada una cadena en el formato "hh:mm:ss", siendo hh el número de horas (entre 00 y 23), mm el número de minutos (entre 00 y 59), y ss el número de segundos (entre 00 y 59), calcula el número de segundos que han transcurrido desde la medianoche anterior.

````python
# APARTADO A)

# Definimos el valor de referencia de e y definamos un n entero positivo grande

E_REF = 2.7182818284590452353602874713527
n = 100000

# Calculamos la aproximación usando la fórmula (1 + 1/n)^n
e_aprox = (1 + 1/n)**n

# Calculamos el error absoluto
error = abs(E_REF - e_aprox)

objetivo_cumplido = error < 10**-4
print("¿El error es menor a 10^-4?:", objetivo_cumplido)
print("El valor de e aproximado que hemos calculado es:", e_aprox)
````

````python
# APARTADO B)

# Definimos el valor de referencia de pi
PI_REF = 3.14159265359

# Calculamos la suma de los primeros 10 términos directamente usando la aproximación de Mashava-Leibniz
suma_serie = (1 - 1/3 + 1/5 - 1/7 + 1/9 - 1/11 + 1/13 - 1/15 + 1/17 - 1/19)
pi_aprox = 4 * suma_serie

# Calculamos el error absoluto
error = abs(PI_REF - pi_aprox)

cumple_requisito = error < 0.1
print("¿El error es menor a 0.1?:", cumple_requisito)
print("El valor de pi aproximado que hemos caculado es:", pi_aprox)

# En este ejercicio he ido añadiendo términos a la suma de la serie hasta alcanzar una error de aproximación absoluto menor que 0.1
````

````python
# APARTADO C)

import math

x = float(input("Introduzca un número en coma flotante: "))

# Basta tener cuidado con el orden y prioridad de operaciones y definir correctamente las funciones de la librería math

m = math.floor((abs(x))/(10**(math.floor(math.log10(abs(x))))))

print("El primer dígito significativo de", x, "es:", m)
````

````python
# APARTADO D)

cadena_horario = input("Introduzca una hora en formato hh:mm:ss: ")

# Para extraer las variables basta ver donde están colocadas en la cadena

horas = int(cadena_horario[:2])
minutos = int(cadena_horario[3:5])
segundos = int(cadena_horario[6:])

total_segundos = horas*3600 + minutos*60 + segundos

print("El total de segundos que han transcurrido desde medianoche es de:", total_segundos, "segundos.")
````

5. Vamos a realizar algunas operaciones con cadenas de caracteres:
    
    a) Lee por teclado una cadena y comprueba si es capicúa, es decir, si se lee igual de izquierda a derecha que de derecha a izquierda. Imprime por pantalla el resultado de la comprobación.

    b) Lee por teclado una cadena y cuenta el número de aes que contiene. Debes tener en cuenta las cuatro formas en que puede aparecer: minúscula sin tilde (“a”), minúscula con tilde (“á”), mayúscula sin tilde (“A”) y mayúscula con tilde (“Á”).

    c) Lee por teclado una cadena y obtén las dos siguientes subcadenas:
    - La subcadena formada por los caracteres en posiciones pares de la cadena original.
    - La subcadena formada por los caracteres en posiciones impares de la cadena original.
    A continuación, imprime por pantalla la longitud de la cadena original, y la suma de las longitudes de las dos subcadenas. Comprueba que ambos números coinciden.
    
    d) Dada una dirección de email con el formato “usuario@dominio”, obtén la subcadena que representa el nombre de usuario, y la subcadena que representa el dominio. Imprime ambas subcadenas por pantalla.

````python
# APARTADO A)

cadena_cap = input("Teclee una cadena de texto: ")

# Basta con saber que la inversa de una cadena se puede ver como cadena[::-1]
# y comparamos caracter a caracter con nuestra cadena inicial
is_capicua = cadena_cap == cadena_cap[::-1]

print("¿Es la cadena", cadena_cap, "capicúa?", is_capicua)
````

````python
# APARTADO B)

cadena_aes = input ("Teclee una cadena de texto: ")

# Usamos la función count() para cada una de las posibles tipos de aes y sumamos sus repeticiones

total_aes = cadena_aes.count("a") + cadena_aes.count("á") + cadena_aes.count("A") + cadena_aes.count("Á")

print("El número total de aes que contiene", cadena_aes, "es:", total_aes)

# Otra forma habría sido pasar la cadena a mayúsculas y únicamente buscar las repeticiones de "A" y "Á".
````

````python
# APARTADO C)

cadena_par_impar = input("Teclee una cadena de texto: ")

# Para la obtención de las cadenas par e impar basta acordarnos de la posibilidad
# de usar step = 2 en la obtención de las subcadenas, una comenzando en 0 y la otra en 1.

subcadena_par = cadena_par_impar[::2]
subcadena_impar = cadena_par_impar[1::2]

print("La subcadena formada por los caracteres en posiciones pares de la cadena original es:", subcadena_par)
print("La subcadena formada por los caracteres en posiciones impares de la cadena original es:", subcadena_impar)

print("La longitud de la cadena original es:", len(cadena_par_impar), "caracteres.")
print("La suma de las longitudes de las dos subcadenas es:", len(subcadena_par) + len(subcadena_impar), "caracteres.")

comprobacion_longitudes = len(cadena_par_impar) == len(subcadena_par) + len(subcadena_impar)

print(comprobacion_longitudes)
````

````python
# APARTADO D)

email = input("Escriba una dirección de correo electrónico (con formato usuario@dominio): ")

# Usamos la función partition() que nos viene muy bien aquí para separar en tres
# partes una cadena mediante, en este caso, el separador '@'.

nombre_usuario = email.partition("@")[0]
dominio = email.partition("@")[2]

print("El nombre de usuario es", nombre_usuario, "y el dominio es", dominio)
````
