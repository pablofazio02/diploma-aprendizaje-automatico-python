1. Escribe un programa que defina cuatro variables de tipo cadena de las siguientes maneras: con comillas dobles, con comillas simples, con una cadena de varias líneas, y con comillas triples. Después debe imprimir por pantalla dichas variables.

```python
dobles = "Tengo un pensamiento"
simples = 'que no me deja sonreír'
varias_lineas = "Esto\ntiene\nvarias\nlineas"
triples = """Esto
tiene que
acabar"""

print(dobles)
print(simples)
print(varias_lineas)
print(triples)
```

2. Escribe un programa que lea una cadena por teclado y un número entero. A continuación, debe imprimir la cadena en líneas distintas tantas veces como indique el número tecleado.

```python
cadena = input("Teclee una cadena: ")
num = int(input("Teclee un número entero: "))
print(cadena * num)
```

3. Escribe un programa que lea una cadena por teclado y después imprima la cadena tres veces: una con todas las letras en minúscula, otra con todas las letras en mayúscula, y otra con las iniciales de las palabras en mayúscula.

```python
cadena = input("Teclee una cadena de texto: ")

print(cadena.lower())
print(cadena.upper())
print(cadena.title())
```

4. Escribe un programa que lea una cadena por teclado, y a continuación imprima el mensaje \<CADENA> tiene \<n> caracteres, siendo \<CADENA> la cadena en mayúsculas y \<n> el número de caracteres de dicha cadena.

```python
cadena = input("Teclee una cadena de texto: ")
print(cadena.upper(), "tiene", len(cadena), "caracteres.")
```

6. Escribe un programa en Python para invertir una cadena introducida por teclado.

```python
cadena = input("Teclee una cadena de texto: ")
longitud = len(cadena)
print(cadena[longitud::-1])
```

7. Escribe un programa que lea una cadena y una letra en minúscula. A continuación, el programa debe imprimir la cadena pero con todas las apariciones de esa letra en mayúscula.

```python
cadena = input("Teclee una cadena de texto: ")
letra = input("Teclee una letra en minúscula: ")
print(cadena.replace(letra, letra.upper()))
```

8. Escribe un programa que lea una dirección de correo electrónico del teclado, con el formato usuario@dominio.  A continuación, debe imprimir el nombre de usuario en una línea, y el dominio en otra línea.

```python
correo = input("Escriba una dirección de correo electrónico: ")
print(correo.partition('@')[0])
print(correo.partition('@')[2])
```

9. Escriba un programa que lea por teclado una fecha en el formato dd/mm/aaaa. A continuación, debe imprimir en líneas distintas el día, el mes y el año.

```python
fecha = input("Teclee una fecha: ")
print(fecha.split('/')[0])
print(fecha.split('/')[1])
print(fecha.split('/')[2])
```

10. Escriba un programa que lea por teclado una lista de nombres de ciudades, separados por comas. A continuación, debe imprimir cada nombre en una línea distinta.

```python
listado = input("Teclee una lista de nombres de ciudades separados por comas: ")

for ciudad in listado.split(','):
    print(ciudad)
```

11. Escriba un programa que lea por teclado los siguientes datos:
- El nombre de un producto
- El precio del producto por unidad (puede contener decimales)
- El número de unidades de producto (es un número entero).

  A continuación, debe imprimir por pantalla el siguiente texto:
  \<nombre_producto>: \<unidades_producto> unidades x \<precio_unitario>€ = \<total>€ siendo \<unidades_producto> imprimido con cinco dígitos, \<precio_unitario> el precio por unidad con 6 dígitos enteros y 2 decimales y \<total> es el coste total con 8 dígitos enteros y 2 decimales.

```python
nombre_producto = input("Teclee el nombre de un producto: ")
precio_producto = float(input("Teclee el precio del producto por unidad: "))
numero_unidades = int(input("Teclee el número de unidades del producto: "))

total = precio_producto * numero_unidades

print(f'{nombre_producto}: {numero_unidades:5d} unidades x {precio_producto:6.2f}€ = {total:8.2f}€')

```
