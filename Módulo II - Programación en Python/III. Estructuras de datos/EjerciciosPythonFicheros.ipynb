1. Escribe un programa en Python que lea un fichero de texto plano. Cada línea del fichero contiene dos números reales separados por un espacio, que representan las dos coordenadas de un punto sobre el plano. El programa debe calcular la longitud de la línea poligonal que define la secuencia de puntos.

```python
import math
```

```python
longitud_total = 0
punto_anterior = None

with open('/content/puntos.txt', 'r') as f:
    for linea in f:
      datos = linea.strip().split()

      x = float(datos[0])
      y = float(datos[1])
      punto_actual = (x, y)

      if punto_anterior is not None:
          dx = punto_actual[0] - punto_anterior[0]
          dy = punto_actual[1] - punto_anterior[1]

          distancia = math.hypot(dx, dy)
          longitud_total += distancia

      punto_anterior = punto_actual

print(f"La longitud total de la línea poligonal es {longitud_total}")
```

2. Escribe un programa en Python para leer un fichero de texto plano. El programa debe contar el número de veces que aparece cada palabra que ocurra en el fichero, y almacenar esta información en un diccionario. Las claves son las palabras, y los valores los recuentos.

```python
dictCont = {}

with open('/content/contar.txt') as f:
  for line in f:
    line = line.lower().replace(',', '').replace('.', '')
    lista_linea = line.strip().split()
    for elem in lista_linea:
      if elem in dictCont:
        dictCont[elem] += 1
      else:
        dictCont[elem] = 1

print(dictCont)
```

3. Escribe un programa en Python para leer un fichero de texto plano que contiene números enteros separados por espacios. El programa debe escribir un nuevo fichero de texto plano que contenga los mismos números, pero en orden inverso.


```python
f = open('/content/enteros.txt')

lista_codigo = []

for line in f:
  lista_codigo = line.split()

lista_codigo_inv = lista_codigo[::-1]

with open('/content/enterosInv.txt', 'w') as fw:
  for elem in lista_codigo_inv:
    fw.write(elem + " ")
```

4. Escribe un programa en Python que lea un fichero de texto plano. Cada línea del fichero contiene el nombre de una localidad seguido de una coma, después el nombre de una provincia seguido de una coma, un código postal de cinco dígitos, y por último la población de dicha localidad expresada como número natural. El programa debe cargar toda esta información en un diccionario cuyas claves sean los nombres de las localidades, y los valores sean tuplas con el resto de los datos leídos.

```python
with open('/content/localidades.txt') as f:
  diccionario_loc = {}
  for line in f:
    datos = line.split(sep = ", ")
    localidad = datos[0]
    provincia = datos[1]
    codigo_postal = int(datos[2])
    poblacion = int(datos[3])
    diccionario_loc[localidad] = (provincia, codigo_postal, poblacion)

print(diccionario_loc)
```

5. Escribe un programa en Python que lea un fichero de texto plano. Cada línea del fichero contiene el nombre de pila de una persona seguido de una coma, después su primer apellido seguido de una coma, a continuación, su segundo apellido seguido de una coma, y por último su fecha de nacimiento en el formato DD-MM-AAAA (día-mes-año). El nombre de pila y los apellidos pueden contener espacios en blanco, que deben conservarse. El programa debe cargar toda esta información en una lista con un elemento por persona. Cada elemento de la lista debe ser una tupla con la información de una persona.

```python
lista_res = []

with open('/content/nombres.txt') as f:
  for line in f:
    lista_datos = line.strip().split(sep = ",")
    nombre = lista_datos[0]
    apellido1 = lista_datos[1]
    apellido2 = lista_datos[2]
    fecha_nac = lista_datos[3]
    lista_res.append((nombre, apellido1, apellido2, fecha_nac))

print(lista_res)
```
