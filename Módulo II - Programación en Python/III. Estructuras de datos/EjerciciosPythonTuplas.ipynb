1. Escribe un programa que, dada una lista de tuplas, las ordena en orden descendente.

2. Escribe un programa que, dada una lista de tuplas, calcule la suma de los primeros y terceros elementos de cada tupla, almacenando ambas sumas en una nueva tupla.

```python
lista = [(3, 4, 51), (19,2,23), (2,2,2)]
suma1 = 0
suma3 = 0

for i in range(len(lista)):
    suma1 += lista[i][0]
    suma3 += lista[i][2]

resultado = (suma1, suma3)
print(resultado)
```

3. Escribe un programa que, dada una lista de tuplas, devuelve una lista nueva que solo contiene las tuplas de longitud estrictamente mayor que uno.

```python
lista = [(3, 4, 51), (19,2,23), (2,2,2), (), (1,)]
lista_res = []

for i in lista:
  if len(i) > 1:
    lista_res.append(i)

print(lista_res)
```

4. Escribe un programa que, dada una lista de tuplas no vacías, compruebe si hay alguna tupla cuyos elementos sean todos iguales.

```python
def son_iguales(tupla_param):
  son_iguales = True

  if len(tupla_param) > 1:
    for j in range(len(tupla_param) - 1):
      if tupla_param[j] != tupla_param[j+1]:
        son_iguales = False
        break

  return son_iguales
```

```python
lista = [(3, 4, 51), (19,2,23), (2,2,2), (1,)]

found = False

for i in lista:
  if (son_iguales(i)):
    found = True
    break

if found:
  print("Existe alguna tupla cuyos elementos son todos iguales.")
else:
  print("No existe ninguna tupla cuyos elementos sean iguales.")
```

5. Escriba un programa que, dadas dos tuplas, devuelva una nueva tupla que solo contenga los elementos comunes a ambas tuplas.

```python
tupla1 = (3,4,5)
tupla2 = (8,9,3)

lista = []

for i in tupla1:
  if i in tupla2:
    lista.append(i)

print(tuple(lista))
```

6. Escribe un programa que, dada una lista de tuplas, calcula una nueva lista que contenga las sumas de los elementos de dichas tuplas y que esté ordenada en orden decreciente.

```python
lista = [(3, 4, 51), (19,2,23), (2,2,2), (1,)]
lista_res = []

for i in lista:
  suma_i = 0
  for j in i:
    suma_i += j
  lista_res.append(suma_i)

print(sorted(lista_res, reverse = True))
```
