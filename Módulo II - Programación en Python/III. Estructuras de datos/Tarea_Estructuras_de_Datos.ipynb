````python
# Inicializamos la lista de entrada

clientes = [
        # Datos edades y gasto por cliente
        [25, 30, 22, 35, 28, 40, 45, 32],
        [35.50, 42.00, 29.75, 52.25, 47.00, 60.30, 55.20, 45.10],

        # Datos género y estado civil
        ["Masculino", "Femenino", "Femenino", "Masculino", "Masculino", "Femenino", "Femenino", "Masculino"],
        ["Casado", "Casado", "Casado", "Soltero", "Soltero", "Soltero", "Soltero", "Casado"],

        # Datos de fecha y hora
        ["2023-08-01", "2023-09-12", "2023-07-22", "2023-09-30", "2023-08-15", "2023-10-02", "2023-06-15", "2023-08-27"],

        # Satisfacción
        ["Muy buena", "Muy mala", "Regular", "Regular", "Muy buena", "Mala", "Regular", "Muy buena"],

        # Nivel educativo
        ["Secundaria", "Universidad", "Doctorado", "Universidad",
        "Secundaria", "Maestría", "Doctorado", "Secundaria"],
]
````

````python
# APARTADO A)

# Inicializamos las variables contadoras del gasto y número de clientes
gasto_total = 0
num_clientes = len(clientes[0])
num_clientes_concretos = 0

for i in range(num_clientes):
  if clientes[0][i] > 30 and clientes[3][i] == "Casado": # Comprobamos que la edad sea mayor que 30 y su estado civil sea 'Casado'
    gasto_total += clientes[1][i]
    num_clientes_concretos += 1

print(f"El gasto medio de las personas casadas mayores de 30 años es de {gasto_total/num_clientes_concretos}.")
````

````python
# APARTADO B)

# Creamos el diccionario como se nos pide, con la satisfacción como clave y la
# puntuación asociada a dicha satisfacción como valor

dic_satisfaccion = {"Muy buena": 5, "Buena" : 4, "Regular" : 3, "Mala" : 2, "Muy mala" : 1}

satisfaccion_total = 0
num_clientes = len(clientes[5]) # Basta ver el tamaño de cualquiera de las sublistas de clientes

for elem in clientes[5]:
  # Para cada clave, obtengo su valor con la función get()
  # NOTA: Si en los datos hubiera alguna clave que no estuviera en el diccionario
  # creado, se salta
  if dic_satisfaccion.get(elem) != None:
    satisfaccion_total += dic_satisfaccion.get(elem)

print(f"La satisfacción media de los clientes es de {satisfaccion_total/num_clientes} sobre 5.")
````

````python
# APARTADO D)

# Basta usar las funciones matemáticas min() y max() sobre un objeto iterable
edad_maxima = max(clientes[0])
edad_minima = min(clientes[0])

print(f"La edad máxima de los clientes es de {edad_maxima} años.")
print(f"La edad mínima de los clientes es de {edad_minima} años.")
````

2. Lee el documento "ventas.txt" que se adjunta en esta tarea. Para ello descarga el fichero, arrástralo a la sección de archivos de tu Google Colab. A continuación lee el fichero mediante código.

    **a)** Crea un diccionario de datos con el contenido del fichero y muestra el diccionario por pantalla.

    **b)** Usando el diccionario, calcula la cantidad de ingresos totales (unidades vendidas x precio_por_unidad).

````python
# APARTADO A)

# Inicializo el diccionario donde guardaremos los contenidos del fichero
diccionario_datos = {}

with open('/content/ventas.txt', 'r') as f:
  for line in f:

    # Esto nos permite limpiar las comas separadoras y dividir las líneas
    datos = line.strip().split(sep = ",")

    # Obtenemos ahora los datos de cada línea
    nombre_dispositivo = datos[0]
    precio_por_unidad = float(datos[1])
    unidades_vendidas = int(datos[2])

    # NOTA: Es importante hacer casting de 'float' e 'int' para luego poder hacer
    # operaciones numéricas con estos datos

    # Insertamos en el diccionario de forma que el nombre del producto sea la
    # clave y la tupla (precio, unidades) sea el valor

    diccionario_datos[nombre_dispositivo] = (precio_por_unidad, unidades_vendidas)

    # NOTA: Si hubiera un dispositivo repetido en la lista, se sobreescriben los
    # antiguos valores

print(diccionario_datos)
````

````python
# APARTADO B)

ingresos_totales = 0

# Basta hacer un bucle por la lista de valores del diccionario, sumando
# el precio*unidad de cada artículo

for (precio, unidad) in diccionario_datos.values():
  ingresos_totales += precio*unidad

print(f"La cantidad de ingresos totales es de {ingresos_totales} euros.")
````

3. Considera los siguientes datos de varias personas, en “X” se almacena el peso y en “Y” el sexo (0: mujer, 1: hombre):

    ```
                                                    X = {55, 80, 78, 92, 65}
                                                      Y = {0, 1, 1, 1, 0}
    ```
    Guárdalos en dos listas y, ayudándote de la función "zip", calcula el peso medio de cada uno de los sexos y muéstralo por pantalla.


````python
# Creamos las listas con los datos de entrada

X = [55, 80, 78, 92, 65]
Y = [0, 1, 1, 1, 0]

# Inicializamos las variables que nos permiten llevar el conteo
cont_M = 0
cont_H = 0
suma_pesos_M = 0
suma_pesos_H = 0

# zip() nos permite iterar varias listas por separado

for peso, sexo in zip(X, Y):
  if sexo == 0: # la persona es mujer
    suma_pesos_M += peso
    cont_M += 1
  elif sexo == 1: # la persona es hombre
    suma_pesos_H += peso
    cont_H += 1

print(f"El peso promedio de los hombres es {suma_pesos_H/cont_H} kilos.")
print(f"El peso promedio de las mujeres es {suma_pesos_M/cont_M} kilos.")
````

4. Vamos a trabajar con listas de cadenas generadas aleatoriamente:
    
    **a)** Genera dos listas de longitud aleatoria formadas por cadenas aleatorias. Para ello, elige para cada lista su longitud aleatoriamente entre 0 y 20 elementos. A continuación, para cada cadena de cada lista elige su longitud aleatoriamente entre 0 y 15 caracteres. Después, para cada carácter de cada cadena de cada lista, elige aleatoriamente el carácter entre '`A`' y '`Z`'. Por último, imprime por pantalla ambas listas.
    
    *Pistas: Debes usar la función `randint` del paquete `random`, para ello debes comenzar el script con la siguiente instrucción:*

    `from random import randint`

    *También debes usar las funciones `chr` para convertir de número a carácter, y `ord` para convertir de carácter en número.*

    **b)** Mezcla las dos listas en una única lista ordenada en orden descendente.

    **c)** Convierte a minúsculas las cadenas de la lista ordenada.

    **d)** Calcula, para cada cadena de la lista ordenada, cuántos caracteres hay al principio de dicha cadena que estén ordenados en orden ascendente. Por ejemplo, para la cadena "`hkrbcopq`" el resultado es 3 porque los caracteres ordenados en orden ascendente al principio son "`hkr`". Por último, calcula e imprime por pantalla la media de los resultados obtenidos para todas las cadenas de la lista ordenada.

````python
from random import randint
````

````python
# APARTADO A)

# Elegimos aleatoriamente la longitud de las listas con rangos entre 0 y 20 elementos
len_lista1 = randint(0,20)
len_lista2 = randint(0,20)

# Inicializamos las listas a listas vacías para poder insertar en ellas
lista1 = []
lista2 = []

for i in range(len_lista1):
  # Elegimos para cadena de cada lista su longitud aleatoriamente entre 0 y 15 caracteres
  len_lista1_cad_i = randint(0,15)
  cad_i = ""
  for j in range(len_lista1_cad_i):
    # Elegimos una letra aleatoriamente entre 'A' (Código ASCII 65) y 'Z' (ASCII 90)
    letter_i = chr(randint(65,90))
    cad_i += letter_i
  lista1.append(cad_i)

for i in range(len_lista2):
  # Elegimos para cadena de cada lista su longitud aleatoriamente entre 0 y 15 caracteres
  len_lista2_cad_i = randint(0,15)
  cad_i = ""
  for j in range(len_lista2_cad_i):
    # Elegimos una letra aleatoriamente entre 'A' (Código ASCII 65) y 'Z' (ASCII 90)
    letter_i = chr(randint(65,90))
    cad_i += letter_i
  lista2.append(cad_i)

print(lista1)
print(lista2)
````

````python
# APARTADO B)

# Para combinar las listas basta concatenarlas

lista_combinada = lista1 + lista2

# Para ordenar descendentemente uso la función sorted() y el parámetro reverse = True

lista_combinada_sorted = sorted(lista_combinada, reverse = True)

print(lista_combinada_sorted)
````

````python
# APARTADO C)

# Basta iterar sobre la lista y modificar cada cadena usando la función lower()

for i in range(len(lista_combinada_sorted)):
  lista_combinada_sorted[i] = lista_combinada_sorted[i].lower()

print(lista_combinada_sorted)
````

````python
# APARTADO D)

lista_cont_car = []

# Iterare sobre cada cadena de la lista comenzando por el índice i = 0 y comparando
# con el índice i+1 hasta que encuentre que no tengan orden ascendente o igual

for cadena in lista_combinada_sorted:
  cont_cadena = 0
  for i in range(len(cadena) - 1):
    if cadena[i] <= cadena[i+1]: # ej: 'a' <= 'b' True
      cont_cadena+=1
    else:
      break
  # Creo una lista que irá guardando para cada cadena, el número que nos piden
  lista_cont_car.append(cont_cadena)

# NOTA: Es importante parar (con break) cuando encontramos un substring descendente
# pues se nos pide que desde el inicio, el número de caracteres ordenados ascendentemente
# no podemos continuar el conteo tras esto

print(lista_cont_car)
print(f"La media de caracteres ordenados ascendentemente al comienzo de las cadenas es {sum(lista_cont_car)/len(lista_cont_car)} caracteres.")
````

5. Vamos a trabajar con conjuntos de números enteros generados aleatoriamente:

    **a)** Genera 20 conjuntos de números enteros elegidos al azar. Cada conjunto debe contener números enteros entre 0 y 99, ambos inclusive. Cada número entero entre 0 y 99 tendrá una probabilidad del 10% de pertenecer a cada conjunto. Imprime los 20 conjuntos por pantalla.
    
    *Pista: Debes usar la función `random` del paquete `random`, comienza tu script con la siguiente instrucción:*

    `from random import random`

    **b)** Calcula, para cada par de conjuntos, su conjunto unión. Imprime por pantalla la media del número de elementos de los conjuntos unión obtenidos de esta manera.

    **c)** Imprime en pantalla los números enteros entre 0 y 99 que aparecen en un conjunto y el siguiente.

    **d)** Halla una lista de 100 elementos, siendo cada elemento una tupla (n,m), donde n es un número entre 0 y 99, y m es el número de conjuntos en los que aparece n. Imprime dicha lista por pantalla.

````python
from random import random
````

````python
# APARTADO A)

# Inicializamos una lista que contendrá los 20 conjuntos para poder posteriormente
# hacer operacioness con ellos
lista_conjuntos = []

for i in range(0, 20):
  # Inicializamos un conjunto vacío con set()
  conjunto = set()

  # Iteramos a través de todos los números enteros del 0 al 99
  for numero in range(100):
    # random() genera un valor aleatorio entre 0.0 y 1.0, luego si cae en menos
    # de 0.10, añadimos dicho número al conjunto (representamos el 10% de probabilidad)
    if random() <= 0.10:
      conjunto.add(numero)

  lista_conjuntos.append(conjunto)
  print(f"Conjunto {i+1}: {conjunto}")
````

````python
# APARTADO B)

suma_total_elementos = 0
numero_de_pares = 0

# Es importante notar que debemos hacer todas las posibles combinaciones de conjuntos
# sin repetir ninguna unión, el primer bucle va de los índices 0 a 19 y el segundo
# desde i + 1 a 19, de forma que nunca repitamos una combinación

for i in range(len(lista_conjuntos)):
    for j in range(i + 1, len(lista_conjuntos)):

      set_1 = set(lista_conjuntos[i])
      set_2 = set(lista_conjuntos[j])

      union = set_1 | set_2

    suma_total_elementos += len(union)
    numero_de_pares += 1

print(f"La media de elementos en los conjuntos unión es: {suma_total_elementos/numero_de_pares:.2f}")
````

````python
# APARTADO C)

# Iteramos sobre la lista de conjuntos y comparamos los elementos del conjunto i
# con los elementos del conjunto i+1, guardandolos en una lista auxiliar

for i in range(len(lista_conjuntos) - 1):
  lista_iguales = []
  for elem in lista_conjuntos[i]:
    if elem in lista_conjuntos[i+1]:
      lista_iguales.append(elem)

  print(f"Coincidentes entre el conjunto {i} y el conjunto {i+1}: {lista_iguales}")
````

````python
# APARTADO D)

# He tenido la idea de guardar de forma más fácil estos datos en un diccionario
# y luego imprimir por pantalla la lista items()

dict_contadora = {}

for elem_set in lista_conjuntos:
  for elem in elem_set:
    if elem not in dict_contadora.keys():
      dict_contadora[elem] = 1
    else:
      dict_contadora[elem] += 1

# A aquellos valores que no han aparecido en el conteo, les asignamos cero repeticiones
for i in range(100):
  if i not in dict_contadora.keys():
    dict_contadora[i] = 0

print(dict_contadora.items())
````
