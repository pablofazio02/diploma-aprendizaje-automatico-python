1. Escribe un programa que, dada una lista de tuplas, calcule un diccionario en el que las claves son los primeros elementos de las tuplas, y los valores son los segundos elementos de dichas tuplas.

```python
tuples_lista = [(1, 'María'), (2, 'Antonio'), (3, 'Juan'), (4, 'Santiago')]

diccionario_tuplas = dict(tuples_lista)

print(diccionario_tuplas)
```

2. Escribe un programa que encuentra el máximo valor en un diccionario e imprime la clave correspondiente.

```python
dict_values = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

for key, value in dict_values.items():
  if value == max(dict_values.values()):
    print(key)
    break
```

3. Escribe un programa que encuentra la mínima clave en un diccionario e imprime el valor.


```python
dict_min_keys = {33 : 'Málaga', 47:'Córdoba', 1:'Madrid', 10:'Barcelona'}

for keys, values in dict_min_keys.items():
  if keys == min(dict_min_keys.keys()):
    print(values)
    break
```

4. Escriba un programa que convierte una tupla de diccionarios en un solo diccionario combinando los diccionarios.

```python
tupla_diccionarios = ({'María': 5, 'Celia' : 9}, {'Antonio': 7, 'Pablo': 10, 'Raúl' : 3, 'Andrés' : 4, 'Celia' : 3, 'Marcos' : 3})

dict1, dict2 = tupla_diccionarios

dict_combinado = dict1 | dict2

print(dict_combinado)
```

5. Escriba un programa que compruebe si dos diccionarios tienen los mismos pares clave-valor.

```python
diccionario1 = {'a': 12, 'b': 10, 'c': 13, 'd': 12, 'e': 7}
diccionario2 = {'c': 13, 'e': 7, 'd': 12, 'a': 12, 'b': 10}

iguales = True

for (keys1, values1) in diccionario1.items():
  igualdad_key = False
  for (keys2, values2) in diccionario2.items():
    if keys1 == keys2 and values1 == values2:
      igualdad_key = True

  if (not igualdad_key):
     iguales = False
     break

if (iguales):
  print("Los diccionarios tienen los mismos pares clave-valor.")
else:
  print("Los diccionarios no tienen los mismos pares.")
```

6. Escribe un programa que, dado un diccionario y un valor, cree un nuevo diccionario con todos los ítems del diccionario de entrada salvo los que contengan ese valor.

```python
diccionario_ejemplo = {'a': 12, 'b': 10, 'c': 13, 'd': 12, 'e': 7}
valor = 12

dict_nuevo = {keys : value for (keys,value) in diccionario_ejemplo.items() if value != valor}
print(dict_nuevo)
```

7. Escribe un programa que, dado un diccionario, calcule una lista con las claves asociadas al valor mínimo del diccionario.

```python
diccionario_min = {'Antonio': 7, 'Pablo': 10, 'Raúl' : 3, 'Andrés' : 4, 'Celia' : 3, 'Marcos' : 3}

minimo_valor = min(diccionario_min.values())

lista_keys_min = [keys for (keys,values) in diccionario_min.items() if values == minimo_valor]

print(lista_keys_min)
```

8. Escribe un programa que, dada una cadena, cuente la frecuencia con la que cada carácter aparece en la cadena, y almacene el resultado en un diccionario.

```python
cadena = "ecdhuec dwbsxijbuuqs"

dict_contar = {}

for i in cadena:
  if i in dict_contar.keys():
    dict_contar[i] += 1
  else:
    dict_contar[i] = 1

print(dict_contar)
```
