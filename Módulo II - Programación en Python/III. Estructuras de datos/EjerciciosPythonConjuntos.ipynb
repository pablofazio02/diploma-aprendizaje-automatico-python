1. Escribe un programa en Python que cree un conjunto vacío, y a continuación inserte todos los números enteros divisibles por 3 y menores que 100.

```python
conjunto_res = set()

for i in range(1,100):
  if i % 3 == 0:
    conjunto_res.add(i)

print(conjunto_res)
```

2. Escribe un programa en Python que calcule la unión del conjunto de los números naturales menores que 100 que son divisibles por 7 con el conjunto de los números naturales menores que 200 que son divisibles por 13.

```python
A = {num for num in range(1,100) if num % 7 == 0}

B = {num for num in range(1,200) if num % 13 == 0}

print(A | B)
```

3. Escribe un programa que elimine los elementos duplicados de una lista usando un conjunto.

```python
lista = [1,13,45,3,4,8,3,2,8,4,9,1,2,2]

lista_set = set(lista)

lista = list(lista_set)

print(lista)
```

4. Escribe un programa en Python para contar el número de caracteres diferentes que ocurren en una cadena introducida por teclado.

```python
cadena = "bedchujxiwskxijcbed"

conjunto_cadena = set(cadena)

print(len(conjunto_cadena))
```

5. Escribe un programa en Python que, dadas dos cadenas, imprima por pantalla los caracteres diferentes que aparecen en la primera cadena pero no en la segunda.

```python
cadena1 = "hvduxhbs"
cadena2 = "jndcjwuisq"

for car in set(cadena1):
  if car not in set(cadena2):
    print(car)
```

6. Escribe un programa en Python que, dadas tres listas de números enteros, imprima por pantalla aquellos números que aparezcan al menos una vez en cada una de las tres listas.

```python
lista1 = [1,53,8,2,6,609,35,29,100,34]
lista2 = [8,3,7,45,68,32,468,609]
lista3 = [7,98,76,38,21,678,345,8,609]

lista1_set = set(lista1)
lista2_set = set(lista2)
lista3_set = set(lista3)

for num in lista1_set:
  if num in lista2_set and num in lista3_set:
    print(num)
```
