3. Escribe un programa que pregunte por teclado un número entero entre el 1 y el 7, ambos inclusive. A continuación, debe imprimir un mensaje mostrando un día de la semana según dicho número: si se escribió el 1, debe imprimir *“LUNES”*, si se escribió el 2, debe imprimir *“MARTES”*, y así sucesivamente hasta el 7 (*“DOMINGO”*). Si se introdujo un número que no esté entre el 1 y el 7, debe imprimir un mensaje de error.

```python
numero = int(input("Teclee un número entero entre el 1 y el 7 (inclusive): "))

match numero:
  case 1:
    print("LUNES")
  case 2:
    print("MARTES")
  case 3:
    print("MIÉRCOLES")
  case 4:
    print("JUEVES")
  case 5:
    print("VIERNES")
  case 6:
    print("SÁBADO")
  case 7:
    print("DOMINGO")
  case _:
    print("ERROR: número no válido.")
```

4. Escribe un programa que pregunte por teclado un número entero. A continuación, debe imprimir uno de estos mensajes según el caso: “Es positivo par”, “Es positivo impar”, “Es negativo par”, “Es negativo impar”, “Es cero”.

```python
numero = int(input("Teclee un número entero: "))

es_par = numero % 2 == 0
es_positivo = numero > 0
es_negativo = numero < 0

if es_par and es_positivo:
  print("Es positivo par")
elif es_par and es_negativo:
  print("Es negativo par")
elif es_positivo and not(es_par):
  print("Es positivo impar")
elif es_negativo and not(es_par):
  print("Es negativo impar")
else:
  print("Es cero")
```

6. Escribe un programa en Python que compruebe si una cadena introducida por teclado es un palíndromo.

11. Escribe un programa en Python que lea un número entero positivo por teclado. Después debe imprimir un mensaje indicando si ese número es primo.

```python
num = int(input("Introduce un número entero positivo por teclado: "))

es_primo = True

for i in range(2, num):
  if (num % i == 0):
    es_primo = False
    break

if es_primo:
  print("El número", num, "es primo.")
else:
  print("El número", num, "no es primo.")
```

12. Escribe un programa en Python que lea un número entero n por teclado. A continuación, debe imprimir el primer número entero mayor o igual que n que sea múltiplo de 13.

```python
num = int(input("Introduce un número entero por teclado: "))

i = num

while True:

  if (i % 13 == 0):
      num_found = i
      break

  i += 1

print("El número encontrado es", num_found)
```

14. Escribe un programa en Python para eliminar todas las vocales de una cadena introducida por teclado.

```python
cadena = input("Teclee una cadena de texto: ")

cadena_res = ""

for letter in cadena:
  if letter not in "AEIOUaeiouáéíóúÁÉÍÓÚ":
    cadena_res += letter

print("La cadena con vocales eliminadas es:", cadena_res)
```

15. Escriba un programa para imprimir todos los números primos entre a y b, siendo a y b números enteros introducidos por teclado.


```python
from sympy import isprime

a = int(input("Introduce un número entero por teclado: "))
b = int(input("Introduce un número entero por teclado: "))

for j in range(a+1, b):
  if isprime(j):
    print(j, end = " ")
```

16. Escriba un programa para generar la sucesión de Fibonacci hasta un número de términos introducido por teclado.

```python
num = int(input("Introduce un número entero positivo por teclado: "))

a = 0
b = 1
next = b
count = 1

while count <= num:
    print(next, end=" ")
    count += 1
    a, b = b, next
    next = a + b
```
