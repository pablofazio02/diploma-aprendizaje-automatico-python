1. Tenemos cadenas de texto compuestas por letras y dígitos. Podrá ser de cualquier longitud par y mayor o igual que dos. Cada cadena estará formada por pares *{letra, dígito}*, de manera que después de cada letra habrá un único dígito, por ejemplo:

    ```
                                                        Z3I4D3Y3H3P1
                                                          A2I0D3B1
    ```
    Escribe un script que procese cualquier cadena como la descrita y escriba por pantalla cada letra repetida tantas veces como indique el dígito que le acompaña. Por ejemplo, para la entrada `A2I0D3B1`, la salida será:

    ```
    A
    A
    D
    D
    D
    B
    ```



````python
cadena_ejemplo = "A2I0D3B1"

# Asumimos que no es necesario comprobar que la cadena de texto
# es de longitud par con pares {letra, dígito}

# De esta forma, podemos recorrer la cadena con un bucle for de step = 2
# y obtener los pares asociados. Para cada par, construimos un bucle anidado
# para imprimir tantas repeticiones de la letra como el dígito nos indique.

for j in range(0, len(cadena_ejemplo), 2):
  letra = cadena_ejemplo[j]
  repeticiones = int(cadena_ejemplo[j+1])

  for _ in range(repeticiones):
    print(letra)
````

2. Crea un script que lea una cadena de texto cualquiera e indique dónde habría que haber escrito mayúscula después de un punto y seguido. Por ejemplo, para la siguiente cadena:

    ```
    esto es un texto de ejemplo. queremos detectar la mayúscula inicial. vamos a usar un script
    en Python. Esta frase no debería marcarse. pero esta sí.
    ```

    La salida debería ser:
    ```
    Error en carácter 29
    Error en carácter 69
    Error en carácter 135
    ```




````python
cadena = """esto es un texto de ejemplo. queremos detectar la mayúscula inicial. vamos a usar un script
en Python. Esta frase no debería marcarse. pero esta sí. """

# Asumo que la cadena se pasa descrita en la anterior variable y no por teclado

# Lo que haremos sea crear una variable booleana que nos indique si hemos visto
# un punto y entonces comprobamos que el siguiente carácter de texto sea una
# minuscúla y así lanzamos el error.

hay_punto = False

for i in range(len(cadena)):
    if cadena[i] == '.':
        hay_punto = True
    elif hay_punto and cadena[i].isspace():  # Importante: el espacio vacío no indica nada
        continue
    elif hay_punto:
        if cadena[i].islower():
            print(f"Error en carácter {i}")
        hay_punto = False
````

3. Aproxima el valor del número pi. Para ello, utiliza la aproximación de Madhava-Leibniz: $$ \pi \sim 4 \left(1 - \dfrac{1}{3} + \dfrac{1}{5} - \dfrac{1}{7} + ...\right) $$

    Debes escribir un script que lea por teclado el error máximo admisible en la aproximación. Después usa la fórmula anterior y ajusta dinámicamente la cantidad de sumandos para que el valor absoluto del error de la aproximación de pi que consigas sea inferior al indicado por teclado. Puedes tomar como valor de referencia pi = 3.14159265359.

    El script deberá mostrar un mensaje cada 1000000 iteraciones con el error de la aproximación en ese momento. Finalmente mostrará dos mensajes, el primero indicando el valor de la aproximación alcanzado, la iteración y el error de aproximación, el segundo mensaje deberá indicar si se ha alcanzado o no el error máximo admisible.
    Nota: Recuerda que debes establecer un límite máximo de iteraciones, por ejemplo, 20000000.

    **Ejemplo de entrada:**

    ```Escribe el error máximo admisible: 0.0000001```

    **Salida:**
    ```
    Iteración: 1000000 | Error: 0.0000010000
    Iteración: 2000000 | Error: 0.0000005000
    Iteración: 3000000 | Error: 0.0000003333
    Iteración: 4000000 | Error: 0.0000002500
    Iteración: 5000000 | Error: 0.0000002000
    Iteración: 6000000 | Error: 0.0000001667
    Iteración: 7000000 | Error: 0.0000001429
    Iteración: 8000000 | Error: 0.0000001250
    Iteración: 9000000 | Error: 0.0000001111
    Iteración: 10000000 | Error: 0.0000001000
    El valor 3.1415927535897814 se ha alcanzado en la iteración 10000000 siendo el error de
    aproximación de: 0.0000001000
    ¡Error absoluto máximo admisible alcanzado! :)
    ```

````python
# Definimos las constantes del enunciado

PI_REF = 3.14159265359
MAX_ITER = 20000000

# Permitimos que le usuario introduzca un error máximo por teclado

error_max = float(input("Escribe el error máximo admisible: "))

error = 1.0
valor = 0.0
i = 0

# Construimos un bucle while que se rompe únicamente al alcanzar el número máximo
# de iteraciones (no obtenemos resultado admisible para el error dado) o cuando
# el error que calculamos sea menor o igual que el requerido

while error > error_max and i < MAX_ITER:

  # Como la fórmula es 4*(1 - (1/3) + (1/5) - (1/7) + ...) en cada iteración
  # añadimos la fracción 1/(2*i + 1). Además, controlamos el cambio de signo
  # viendo si la iteración es un número par o impar (en los pares, sumamos y en
  # los impares, restamos dicha fracción)

  denominador = 2 * i + 1

  termino = 4 / denominador

  if i % 2 == 0:
      valor += termino
  else:
      valor -= termino

  error = abs(valor - PI_REF)

  # Cada 1000000 iteración, imprimos los resultados actuales

  if (i % 1000000 == 0 and i > 0):
    print(f"Iteración: {i} | Error: {error}")

  # Si alcanzamos nuestro objetivo dentro del bucle, lo imprimos por pantalla
  if (error_max >= error):
    print(f"El valor {valor} se ha alcanzado en la iteración {i} siendo el error de aproximación de: {error_max}")
    print("¡Error absoluto máximo admisible alcanzado! :)")

  i += 1

# Si salimos del bucle y la condición de máximo número de iteraciones se ha alcanzado,
# eso significa que hemos salido del bucle while sin obtener un valor con
# error máximo admisible

if(MAX_ITER <= i):
  print("¡Error absoluto máximo admisible no alcanzado! :(")
````

4. Crea un script en Python que pida al usuario calificaciones de los alumnos, sólo se admitirán valores enteros positivos de 0 a 10. Si se introduce un valor superior a 10, se ignorará y se deberá introducir otro valor. Terminará de pedir calificaciones cuando se introduzca un texto no numérico. Debes calcular la nota media, la nota máxima y la mínima.

    a) ¿Has controlado el caso de no introducir ninguna nota válida?, por ejemplo, prueba a introducir:
    ```
    a
    2

    ```
    ¿se produce algún error?, ¿cómo lo has solucionado o solucionarías? Razona tus respuestas.

    b) Prueba a introducir:

    y copia el resultado.

````python
cadena = input("Introduce un número entero positivo (del 0 al 10): ")

# Definimos variables auxiliares que nos ayudarán a controlar las variables de salida
nota_max = -1
nota_min = 11
suma_notas = 0
num_notas = 0

while cadena.isnumeric(): # Mientras lo tecleado por pantalla sea un número

  cadena = int(cadena)

  if 0 <= cadena <= 10: # Comprobamos que es además un número del 0 al 10
    if nota_max < cadena:
      nota_max = cadena
    if nota_min > cadena:
      nota_min = cadena
    suma_notas += cadena
    num_notas += 1
  cadena = input("Introduce un número entero positivo (del 0 al 10): ")

print()

if(num_notas > 0):
  print("La nota media de los alumnos es:", suma_notas/num_notas)
else:
  print("No ha introducido ninguna nota válida.")

if 0 <= nota_max <= 10:
  print("Nota máxima de los alumnos:", nota_max)

if 0 <= nota_min <= 10:
  print("Nota mínima de los alumnos:", nota_min)
````

5. Desarrolla un programa para resolver ecuaciones del tipo $f(x) = 0$, siendo $f$ una función continua. Para ello debes basarte en el teorema de Bolzano, que establece que, siendo $a < b$, si $f(a)$ es de signo opuesto a $f(b)$, entonces existe una solución de la mencionada ecuación entre $a$ y $b$. Partiendo de dos valores $a$, $b$ que cumplan que $f(a)$ es de signo opuesto a $f(b)$, debes considerar el punto medio $m$ entre $a$ y $b$. Si $f(m)$ es de signo opuesto a $f(a)$, entonces debe haber una solución entre $a$ y $m$. En otro caso,
si $f(m)$ es de signo opuesto a $f(b)$, entonces debe haber una solución entre $m$ y $b$. Debes continuar el proceso hasta que el valor absoluto de $f(m)$ sea menor que una tolerancia establecida previamente. Como ejemplo debes considerar: $a=0.0$, $b=1.0$, $tolerancia=1.0e-5$,
$$f(x) = x^3 + 4x^2 - 7x + 1$$
Tu programa debe imprimir los valores de $a$, $b$, $m$, y el valor absoluto de $f(m)$, para cada iteración ejecutada, hasta alcanzar la precisión establecida.

````python
# Definimos las variables de entrada del ejemplo

def f(x):
  return x**3 + 4*x**2 - 7*x + 1

a = 0.0
b = 1.0
tolerancia = 1.0e-5

# Comprobar que f(a) y f(b) tienen signos opuestos para poder aplicar el
# algoritmo

if f(a)*f(b) > 0:
  print("No es posible aplicar el teorema de Bolzano para a y b")
else:
  # Esta forma de uso de 'f-strings' nos sirve para imprimir una tabla más limpia y ordenada
  print(f"{'Iter':<5} {'a':<12} {'b':<12} {'m':<12} {'|f(m)|':<12}")

  iteracion = 1

  # Aplicamos el desarrollo del teorema de Bolzano del enunciado y únicamente
  # paramos cuando obtengamos solución con tolerancia menor a la indicada

  while True:

      m = (a + b) / 2

      f_m = f(m)
      f_a = f(a)

      print(f"{iteracion:<5} {a:<12.6f} {b:<12.6f} {m:<12.6f} {abs(f_m):<12.6f}")

      if abs(f_m) < tolerancia:
          print(f"\nRaíz encontrada: x = {m:.6f}")
          break

      if f_a * f_m < 0:
          b = m
      else:
          a = m

      iteracion += 1
````

6. Escribe un programa que, dada una cadena, halle su subcadena más larga formada por letras mayúsculas. Una vez hallada, debe imprimir dicha subcadena y su longitud. Por ejemplo, para la cadena “*dddSRHOW4775/DFsTY67*”, la subcadena obtenida debe ser “*SRHOW*”, cuya longitud es 5.

````python
cadena = "*dddSRHOW4775/DFsTY67*"

# Variables auxiliares que usamos para ir controlando la subcadena con longitud
# máxima y la subcadena con mayúsculas actual

max_long = 0
cont = 0
max_subcadena = ""
sub = ""

# Iteramos por toda la cadena de letra en letra y si verificamos que es una
# mayúscula, tratamos de construir una subcadena de ellas hasta que encontremos
# un caracter diferente. Por último, comparamos la subcadena creada con la de
# mayor longitud hasta el momento.

for letter in cadena:
  if letter.isupper():
    cont += 1
    sub += letter
  else:
    if cont > max_long:
      max_long = cont
      max_subcadena = sub

    cont = 0
    sub = ""

print("Dada la cadena", cadena, "la mayor subcadena obtenida es", max_subcadena, "de longitud", max_long)
````
