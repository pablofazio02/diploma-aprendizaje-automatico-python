6. Escriba un programa que le pregunte al usuario su peso (en kilogramos) y su talla (en metros), y calcule e imprima su Índice de Masa Corporal (IMC) redondeado a dos decimales. La fórmula que debe usarse es: $$IMC=\frac{peso}{talla^{2}}$$

```python
peso_usuario = float(input("Teclee su peso (en kilogramos): "))
talla_usuario = float(input("Teclee su talla (en metros): "))
imc_usuario = peso_usuario/(talla_usuario**2)
print("Su IMC es:", round(imc_usuario,2))
```

7. Escriba un programa que le pregunte al usuario dos números enteros, que llamaremos dividendo y divisor, respectivamente. A continuación, el programa debe calcular el cociente y el resto de la división entera del dividendo entre el divisor, e imprimir el mensaje: <dividendo> dividido entre <divisor> da cociente <c> y resto <r>

```python
dividendo = int(input("Teclee un primer número entero (dividendo): "))
divisor = int(input("Teclee un segundo número entero (divisor): "))
c = dividendo // divisor
r = dividendo % divisor
print(dividendo, "dividido entre", divisor, "da cociente", c, "y resto", r)
```

8. Escribe un programa que le pregunte al usuario el capital inicial de un préstamo (C), el tipo de interés efectivo (i) y el número de periodos de pago (n). A continuación, el programa debe calcular e imprimir la cuota (a) según el sistema de amortización francés de acuerdo con esta fórmula: $$a=\frac{C*i}{1-(\frac{1}{1+i})^{n}}$$

```python
C = float(input("Teclee el capital inicial del prestamo: "))
i = float(input("Teclee el tipo de interés efectivo: "))
n = int(input("Teclee el número de periodos de pago: "))
print("Cuota:", (C*i)/(1 - (1/(1+i))**n))
```
